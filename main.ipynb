```python
import heapq #拿來快速找出一串 list 中最小的那個
from typing import List, Tuple #type是拿來註明變數類別的工具，但對 python 本體而言沒差，他不會看

def huffman_lengths(p: List[float]) -> List[int]: #List 的功用：標示這個函數是從哪裡打到哪裡
    n = len(p)
    if n < 2:
        raise ValueError("Need at least 2 symbols.")

    # Each heap item: (weight, uid, symbol_set)，後面會用 heappop ，他會從 tuple 的第一個變數開始比
    # 小，而uid會越來越大，是為了確保如果真的遇到有三個以上的 symbol 的出現機率是相同的話，還是可以篩出兩個 
    # symbol_set is a tuple of symbol indices that are inside that subtree.
    heap: List[Tuple[float, int, Tuple[int, ...]]] = [] #冒號是告訴我們 heap 的變數類別
    uid = 0
    for i, w in enumerate(p):
        heap.append((float(w), uid, (i,)))
        uid += 1
    heapq.heapify(heap) #把 heap 轉成可以用 heapq 的函式的結構

    lengths = [0] * n

    while len(heap) > 1:
        w1, _, s1 = heapq.heappop(heap)  # heap 第一個變數最小的，存取之後把它從 heap 裡面拿出來
        #_代表我不需要那個變數
        w2, _, s2 = heapq.heappop(heap)  # second smallest

        # Merging two subtrees increases code length of all their leaves by 1
        for i in s1:
            lengths[i] += 1
        for i in s2:
            lengths[i] += 1

        merged = s1 + s2
        heapq.heappush(heap, (w1 + w2, uid, merged))#用過 heapify 之後要加東西進去只能用 heappush
        uid += 1

    return lengths
```

```python
p = [0.21, 0.6, 0.11, 0.04, 0.04]
u = huffman_lengths(p)
print(u)
```
